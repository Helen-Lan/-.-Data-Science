### Заменяем пропущенные значения:

Считаем количество пропущенных значений в датафрейме:

```python
data.isna().sum()
```

Для оценки влияния пропущенных значений на достоверность всего анализа пройдемся по всем столбцам с пропусками:
- `ceiling_height` (высота потолков). Количество пропусков 9195, что составляет 39%. В данном столбце можно заменить пропуски на медиану;
- `floors_total` (всего этажей в доме). Количество пропусков 86, что менее 1%. Удаляем строки с пропущенными значениями;
- `kitchen_area` (площать кухни). Часть значений заполняем условные 2 метра для квартир-студий и для квартир с открытой планировкой. Остальные значения не меняем;
- `balcony` (балкон). Предположем, что отсутствие сведений в данном столбце означает отсутствие балкона, заменим на `0`;
- `locality_name` (населенный пункт). Количество пропущенных значений 49, что менее 1% всех. Критичный показатель, поэтому пропущенные значения удаляем.

Мы изменили все возможные варианты. Удалять данные из-за пропусков мы не будем, так как не знаем как они могут повлиять на общий разброс данных. В дальнейшем мы создадим отдельный датафрейм, в который войдут необходимые нам значения с расчетом разброса данных.


Заменяем пропущенные значения в столбце `ceiling_height` на медиану:

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

Удаляем объявления в которых пропущены сведения о населенном пункте:

```python
data = data.dropna(subset=['locality_name'])
```

Удаляем объявления в которых пропущены сведения о количестве этажей в доме:

```python
data = data.dropna(subset=['floors_total'])
```

Заменяем пропущенные значения в столбце `balcony` на `0`:

```python
data['balcony'] = data['balcony'].fillna(0)
```

Проверяем значения столбца `is_apartment`. Возможно при заполнении объявления данная графа предполагала только один вариант заполнения, т.е. если в столбце указаны только True, соответственно все остальные должны быть False:

```python
data['is_apartment'].value_counts()
```

Теория не подтвердилась, только 50 собственников указали что это апартаменты. Так могло случиться, если заполнение данного поля не обязательное, либо технический сбой при выгрузке. Так как использовать при анализе мы его не будем, оставим его в первоначальном виде.

Проверяем гипотезу, что пропущенные значения площади кухни могут быть у квартир-студий и у квартир с открытой планировкой:

```python
print('Всего квартир-студий:', data[data['studio'] == True].shape[0])
print('Всего квартир с открытой планировкой:', data[data['open_plan'] == True].shape[0])
```

```python
print('Количество пропущенных значений площади кухни в квартирах студиях:', data[data['studio'] == True]['kitchen_area'].isna().sum())
print('Количество пропущенных значений площади кухни в квартир с открытой планировкой:', data[data['open_plan'] == True]['kitchen_area'].isna().sum())
```

Меняем значения с толбце "kitchen_area" в квартирах-студиях на условных 2 метра, так как кухонное пространство все же есть. Указание 0 может негативно отразиться в дальнейшей работе. В дальнейшем мы будем оценивать влияние площади кухни на стоимость квартиры, и нулевые значения будут учитываться в расчетах и снижать среднюю и медиану. Пропущенные значения не влючаются в расчет.

```python
data.loc[data['studio'] == True, 'kitchen_area'] = 2
```

Аналогично поступаем в квартирах с открытой планировкой:

```python
data.loc[data['open_plan'] == True, 'kitchen_area'] = 2
```

Во всех остальных пропусках столбец `kitchen_area` без изменений, пропущенные значения не меняем. 

```python
# check

data_check = data.query('((living_area + kitchen_area + total_area*0.05) < total_area) \
or (kitchen_area.isna()) \
or (living_area.isna())')
```
